```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
n1 = np.array([2,1,3,2])
n2 = np.array([3,5,1,2])
```

```python
# subplot2grid
plt.subplot2grid((2,2),(0,0))
plt.plot(n1,n2)
plt.subplot2grid((2,2),(0,1), rowspan=2)
plt.plot(n1,n2)
plt.subplot2grid((2,2),(1,0))
plt.plot(n1,n2)
```
